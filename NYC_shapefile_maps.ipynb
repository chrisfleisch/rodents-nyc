### Here Is an outline of the map using basemap. Note the resolution field 'f' setting means full, but takes a couple minutes to run so when just playing around better to set to 'l' (low).

```python
fig, ax = plt.subplots(figsize=(10,20))
ny_lon, ny_lat = -73.95, 40.7
boroughs=['Manhattan','Brooklyn','Queens','Bronx','Staten Island']
m = Basemap(resolution='f', # c, l, i, h, f or None
            projection='merc',
            lat_0=ny_lat, lon_0=ny_lon,
            llcrnrlon=-74.24, llcrnrlat=40.47, urcrnrlon=-73.7, urcrnrlat=40.92)
#westlimit=-74.239956; southlimit=40.471927; eastlimit=-73.697713; northlimit=40.91875
m.drawmapboundary(fill_color='#46bcec')
m.fillcontinents(color='#f2f2f2',lake_color='#46bcec')
m.drawcoastlines()
m.readshapefile('C:/Users/adpra/Documents/MIDS/W209_Data_Visualization/Final_Project/rodents-nyc/Borough_Files/nyc_boroughs','boroughs')
m.readshapefile('C:/Users/adpra/Documents/MIDS/W209_Data_Visualization/Final_Project/rodents-nyc/Zipcode_Files/nyc_zip_codes','zip_codes')
plt.annotate('MANHATTAN',(.45,.7),xycoords='axes fraction')
plt.annotate('BRONX',(.65,.85),xycoords='axes fraction')
plt.annotate('QUEENS',(.7,.55),xycoords='axes fraction')
plt.annotate('BROOKLYN',(.46,.39),xycoords='axes fraction')
plt.annotate('STATEN ISLAND',(.1,.25),xycoords='axes fraction')
m
```
